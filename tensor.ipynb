```python
import tensorflow as tf
```

```python
x = tf.constant(4.0)
print(x)
```

```python
# Rank 0
x = tf.constant(4.0, shape=(1,1), dtype=tf.float32)
print(x)
```

```python
# Rank 1
y = tf.constant([2.0, 3.0, 4.0], shape=(3,1), dtype=tf.float32)
print(y)
```

```python
# Rank 2
z = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
print(z)
#
```

```python
x = tf.ones((3,3))
print(x)
```

```python
y = tf.zeros((2,2))
print(y)
#
```

```python
z = tf.eye(3)
print(z)
#
```

```python
x = tf.random.normal((3,3), mean=0.0, stddev=1.0)
print(x)
```

```python
y = tf.random.uniform((3,3), minval=0, maxval=1)
print(y)
#
```

```python
x = tf.range(10)
print(x)
#
```

```python
x = tf.cast(x, dtype=tf.float32)
print(x)
#
```

```python
x = tf.constant([1,2,3])
y = tf.constant([4,5,6])

z = tf.add(x,y)
z = x + y
print(z)
```

```python
z = tf.subtract(x,y)
z = x - y
print(z)
```

```python

z = tf.divide(x,y)
z = x / y
print(z)

```

```python
z = tf.multiply(x,y)
z = x * y
print(z)
```

```python
z = tf.tensordot(x,y, axes=1)
print(z)
```

```python
z = x ** 3
print(z)
```

```python
x = tf.random.normal((2,3)) # 2,3
y = tf.random.normal((3,4)) # 3,4

z = tf.matmul(x,y)
z = x @ y
print(z)
```

```python
x = tf.constant([[1,2,3,4],[5,6,7,8]])
print(x[0])
print(x[:, 0]) # all rows, column 0
print(x[1, :]) # row 1, all columns
print(x[1,1]) # element at 1, 1
```

```python
x = tf.random.normal((2,3))
print(x)
```

```python
x = tf.reshape(x, (3,2))
print(x)
```

```python
x = tf.reshape(x, (-1,2))
print(x)
```

```python
x = tf.reshape(x, (6))
print(x)
```

```python
x = x.numpy()
print(type(x))
```

```python
x = tf.convert_to_tensor(x)
print(type(x))
```

```python
x = tf.constant("Patrick")
print(x)
```

```python
x = tf.constant(["Patrick", "Max", "Mary"])
print(x)
```

```python
b = tf.Variable([[1.0, 2.0, 3.0]])
print(b)
print(type(b))
```
